## **Consider passion_for_eating and frequency to category**
- Convert passion_for_eating and frequency to category เนื่องจากข้อมูลเป็นการให้ความพึงพอใจจากแบบสอบถามตัวเลขที่ให้มาเป็นเพียงสเกลในการประเมินไม่สามารถนำมาคิดค่าทางสถิติได้เลย

```python
result_count_passion = filtered_data['domain'].eq('email.kmutnb.ac.th').groupby(filtered_data['passion_for_eating']).sum().reset_index(name='count_passion')
result_count_passion.index = result_count_passion.index+1
result_count_passion.sort_values(by = 'passion_for_eating')
```

```python
result_count_freq =  filtered_data['domain'].eq('email.kmutnb.ac.th').groupby(filtered_data['frequency']).sum().reset_index(name='count_f')
result_count_freq.sort_values(by = 'frequency')
```

```python
result_count_freq = result_count_freq.drop(0)
result_count_freq
```

### **Concatenate dataframe**

```python
result_count_pandf = pd.concat([result_count_passion,result_count_freq],axis = 1)
result_count_pandf
```

- นำค่าในตารางไปคิดเปอร์เซ็นต์เพื่อที่จะหาว่าความชื่นชอบของกลุ่มตัวอย่างในมอมีความชื่นชอบในการทานเฟรนฟรายและความถี่อยู่ที่กี่เปอร์เซ็นต์

```python
result_count_pandf['percent_count_passion'] = round(result_count_pandf['count_passion']/sum(result_count_pandf['count_passion'])*100,2)
result_count_pandf['percent_count_f'] = round(result_count_pandf['count_f']/sum(result_count_pandf['count_f'])*100,2)
```

- จากกลุ่มตัวอย่างในมอมากกว่า 80 เปอร์เซ็นต์มีความชื่นชอบในการทานเฟรนฟรายแต่มีความถี่ในการซื้อเฟรนฟรายมาทานปานกลางถึงน้อยมากมากกว่า 80 เปอร์เซ็นต์
- การขายในงาน Openhouse 2 วัน อาจจะตอบโจทย์ในการขายมากกว่าการเช่าที่และขายระยะยาวเพราะคนในมอชอบทานเฟรนฟรายแต่มีความถี่ในการซื้อน้อยเมื่อมีร้านใหม่ๆคนอาจจะมาซื้อเยอะแต่เมื่อนานๆคนอาจจะไม่ค่อยซื้อ

```python
def get_value_counts_and_percentage(data, column_name):
    value_counts = filtered_data[column_name].str.split(',').explode().str.strip().value_counts().to_frame('Count')
    value_counts['Percentage'] = round(value_counts['Count'] * 100 / value_counts['Count'].sum(), 2)
    return value_counts
```
